### Regex

````python
import re
````

````python
re.search ('.*?(ball).*?', 'apple123ball456cat')
````

````python
s = 'apple123ball456cat'
````

````python
grp = re.match ('.*?(\d+).*?', 'apple123ball456cat')
````

````python
grp.groups ()
````

````python
re.findall ('.*?(\d).*?', 'apple123ball456cat')
````

````python
re.findall ('[a-z]*?(\d+)[a-z]*?', 'apple123ball456cat')
````

````python
re.findall ('.*?(\d+).*?', 'apple123ball456cat')
````
