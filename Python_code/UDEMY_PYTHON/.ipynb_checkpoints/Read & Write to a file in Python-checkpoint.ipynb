Simple intrest Depends on three factors 1) Principal Amount 2) The No.of Years 3) Rate of Intrest

```python
#Declaring the Variables
PA = input("Please Enter your Principal Amount PA")
Y = input("Please Enter your No.of.Years Y")
RI = input("Please enter your rate of Interest RI")

#Converting the String Values to Int 
PA = int(PA)
Y = int(Y)
RI = int(RI)

#Formula
SimpleIntrest = (PA*Y*RI)/100

print(SimpleIntrest)
```
