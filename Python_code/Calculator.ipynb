Calculator 

```python
class calculator:
    def sum(self,a,b):
        return a+b
    def sub(self,a,b):
        return a-b
    def mul(self,a,b):
        return a*b
    def div(self,a,b):
        return a//b
x=calculator()
print(x.sum(2,4))
print(x.sub(2,4))
print(x.mul(2,4))
print(x.div(2,4))
```
